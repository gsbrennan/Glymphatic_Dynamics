# Testing parametric assumptions all data

The ANOVA test assumes that, the data are normally distributed and the variance
across groups are homogeneous. We can check that with some diagnostic plots.

The normality assumption can be checked by using one of the following two approaches:

Analyzing the ANOVA model residuals to check the normality for all groups together.
This approach is easier and it’s very handy when you have many groups or if there
are few data points per group. Check normality for each group separately.
This approach might be used when you have only a few groups and many data points per group.

For explanation see e.g.
https://rpubs.com/widi123/Anova_CS

```python
# extreme outlier detection

sd = raw_sd_full
ref = raw_ref_full
data= [ref['91'], ref['172'], ref['176'], ref['178'], ref['183'], ref['190'], ref['191'], ref['205'], ref['215'], ref['218'], ref['228'], ref['240'], ref['175']]
datalabels = ['91','172','176','178','183','190','191','205','215','218','228','240','175']

coeff = 3

for i in range(len(data)):
  d = data[i]
  d.index = rois

  # IQR
  Q1 = np.percentile(d, 25, method='midpoint')
  Q3 = np.percentile(d, 75, method='midpoint')
  IQR = Q3 - Q1

  upper=Q3+coeff*IQR
  lower=Q1-coeff*IQR

  print('Patient:', datalabels[i])

  for j in range(len(d)):
    if d[j] > upper or d[j] < lower:
      print(rois[j])
      print(d[j])
    else:
      continue
```

```python
# reshape the dataframe suitable for statsmodels package
df = ref.copy().reset_index(drop=True)
df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=ref_patients_full)
df_melt.columns = ['index', 'patients', 'clearance_value']

# df = sd.copy().reset_index(drop=True)
# df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=sd_patients_full)
# df_melt.columns = ['index', 'patients', 'clearance_value']

model = ols('clearance_value ~ C(patients)', data=df_melt).fit()

# ANOVA table using bioinfokit v1.0.3 or later (it uses wrapper script for anova_lm)
res = stat()
res.anova_stat(df=df_melt, res_var='clearance_value', anova_model='clearance_value ~ C(patients)')
res.anova_summary
```

```python
# Thresholds for outliers
moderate_threshold = 2
severe_threshold = 3

# Total number of residuals
total_residuals = len(res.anova_std_residuals)

# Identify moderate outliers (|residual| > 2)
moderate_outliers = np.abs(res.anova_std_residuals) > moderate_threshold
num_moderate_outliers = np.sum(moderate_outliers)

# Identify severe outliers (|residual| > 3)
severe_outliers = np.abs(res.anova_std_residuals) > severe_threshold
num_severe_outliers = np.sum(severe_outliers)

# Calculate the percentage of outliers
percent_moderate_outliers = (num_moderate_outliers / total_residuals) * 100
percent_severe_outliers = (num_severe_outliers / total_residuals) * 100

# Print the results
print(f"Total residuals: {total_residuals}")
print(f"Number of moderate outliers (|residual| > 2): {num_moderate_outliers} ({percent_moderate_outliers:.2f}%)")
print(f"Number of severe outliers (|residual| > 3): {num_severe_outliers} ({percent_severe_outliers:.2f}%)")

# Normality visual
sm.qqplot(res.anova_std_residuals, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()

# histogram
plt.hist(res.anova_model_out.resid, bins='auto', histtype='bar', ec='k')
plt.xlabel("Residuals")
plt.ylabel('Frequency')
plt.show()
```

```python
# Note : if sample size is greater than 50, the normal QQ plot is preferred because
# at larger sample sizes the Shapiro-Wilk test becomes very sensitive even to a
# minor deviation from normality.

# Normality statistical - method 1
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
```

```python
# Initialize an empty DataFrame to store the results of times it takes to reach 'activation' (5% threshold)
threshold = 0.017
results = []

# Loop over all patients and ROIs to calculate the time to threshold
for pat in ref_patients:
    for r in rois:
        progression_data = df_ref.loc[(pat), r]
        time_to_threshold = progression_data[progression_data >= threshold].index.min()
        results.append({'Patient': pat, 'ROI': r, 'Time_to_Threshold': time_to_threshold})

# Convert the results to a DataFrame
df_results = pd.DataFrame(results)
```

```python
def extract_samples(regions,samples_per_patient):
  dfs = []
  data = []

  # make a list, dfs, holding dataframes of each patient data with columns being the regions
  for pat in ref_patients:
    # print(pat)
    Df = pd.DataFrame()
    for r in regions:
        # print(r)
        D = pd.read_csv('../data/model_outputs/Brennan-Goriely-Model-Solver-Regional-{}_{}_raw-Misfolded-Protein-Concentration.csv'.format(r,pat), names = ['time','{}'.format(r)], nrows=300)
        Df[r] = D[r]
    dfs.append(Df)

  # for each patient make a list of timepoints taken either from choice of distribution
  for k in range(0,len(ref_patients)):
    print(k)
    tps = randint.rvs(math.floor(root1),math.ceil(root2), size=samples_per_patient)

    # extract the row of regional toxic conc values corresponding to the random time points
    for t in tps:
      data.append(dfs[k].iloc[round(t)])

  for i in range(0,len(data)):
    data[i] = data[i].transpose()

  df3 = pd.concat([data[i] for i in range(0,len(data))],axis=1)
  df4 = df3
  df4.columns = ['sample{}'.format(k) for k in range(1,samples_per_patient*len(ref_patients)+1)]

  return df4
```

```python
def zscore_rank(df, quartile, no_of_samples, regions_list):
  inv = []

  for r in range(1,no_of_samples+1):
    df['sample{}_zscore'.format(r)] = stats.zscore(df4['sample{}'.format(r)])

  df5 = df.iloc[:,no_of_samples:]
  print(df5)
  df5.to_csv('zscores.csv')
  qs = df5.quantile(quartile,axis=0)
  involved = qs.mean()
  print(involved)

  df6 = df5.T
  print(df6.columns)

  for jj in regions_list:
    print(jj)
    n = np.sum(df6[jj] > involved)
    inv.append(n/no_of_samples)

  df7 = pd.DataFrame(list(zip(regions_list, inv)), columns = ["Braak regions", "Simulated tau involvement"])
  df8 = df7.sort_values(by ='Simulated tau involvement', ascending = False )

  return df8
```

```python


# Calculate the mean concentration across time
mean_concentration = df_unB.mean(axis=1)

# Generate the x values based on the length of the data
x = np.arange(len(mean_concentration))

# Apply the Savitzky-Golay filter to smooth the data
smoothed_concentration = savgol_filter(mean_concentration, window_length=51, polyorder=3)

# Define the sigmoid function
def sigmoid(x, L, x0, k):
    return L / (1 + np.exp(-k * (x - x0)))

# Initial guess for the parameters [L, x0, k]
initial_guess = [smoothed_concentration.max(), len(x) / 2, 0.01]

# Fit the sigmoid function to the smoothed data
popt, _ = curve_fit(sigmoid, x, smoothed_concentration, p0=initial_guess)

# Extract the fitted parameters
L, x0, k = popt

# Generate the fitted sigmoid curve
fitted_sigmoid = sigmoid(x, L, x0, k)

# Calculate the derivative of the fitted sigmoid
sigmoid_derivative = k * L * np.exp(-k * (x - x0)) / (1 + np.exp(-k * (x - x0)))**2

# Find the index of the maximum gradient (steepest point)
max_derivative_index = np.argmax(sigmoid_derivative)

# Set a threshold (e.g., 10% of the maximum gradient) to define the invasion window
threshold = 0.1 * sigmoid_derivative[max_derivative_index]

# Find the start of the invasion window (going backwards from the steepest point)
start_index = max_derivative_index
while start_index > 0 and sigmoid_derivative[start_index] > threshold:
    start_index -= 1

# Find the end of the invasion window (going forwards from the steepest point)
end_index = max_derivative_index
while end_index < len(x) - 1 and sigmoid_derivative[end_index] > threshold:
    end_index += 1

# Get the time points and fitted sigmoid values at the invasion window limits
start_time = x[start_index]
end_time = x[end_index]
start_value = fitted_sigmoid[start_index]
end_value = fitted_sigmoid[end_index]

# Plot the fitted sigmoid curve
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, smoothed_concentration, linewidth=3, color="black", label='Smoothed Concentration')
ax.plot(x, fitted_sigmoid, color='red', linestyle='--', linewidth=2, label='Fitted Sigmoid')

# Highlight the invasion window
ax.axvspan(start_time, end_time, color='orange', alpha=0.3, label='Invasion Window')

# Highlight the steepest point
ax.scatter(x[max_derivative_index], fitted_sigmoid[max_derivative_index], color='blue', zorder=5, label='Steepest Point')

# Set labels and legend
plt.xlabel('Time', fontsize=14)
plt.ylabel('Misfolded Protein Concentration', fontsize=14)
plt.legend()

# Set limits for x and y axes based on data
plt.xlim([0, len(x)])
plt.ylim([0, smoothed_concentration.max() + 0.05])

# Show the plot
plt.show()

# Print information about the invasion window
print(f"Invasion Window: Start Time = {start_time}, End Time = {end_time}")
print(f"Fitted Sigmoid Value at Start: {start_value}")
print(f"Fitted Sigmoid Value at End: {end_value}")
```

```python
# Example structure: columns are patient IDs, rows represent clearance values over time or ROI
# Assume raw_ref is structured this way

# Reshape the DataFrame to long format for statistical testing
long_format_df = raw_ref.melt(var_name='Patient', value_name='Clearance')

# Check if there are statistically significant differences using the Kruskal-Wallis test first
stat, p_value = kruskal(*[group['Clearance'].dropna().values for _, group in long_format_df.groupby('Patient')])

print(f"Kruskal-Wallis test statistic: {stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Significant differences found between patient groups. Proceeding with pairwise comparisons...")

    # Prepare lists for storing pairwise comparison results
    patients = long_format_df['Patient'].unique()
    p_values = []
    comparisons = []

    # Perform pairwise Mann-Whitney U tests between each pair of patients
    for i, pat1 in enumerate(patients):
        for pat2 in patients[i + 1:]:
            group1 = long_format_df[long_format_df['Patient'] == pat1]['Clearance'].dropna()
            group2 = long_format_df[long_format_df['Patient'] == pat2]['Clearance'].dropna()
            if len(group1) > 0 and len(group2) > 0:
                stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
                p_values.append(p)
                comparisons.append((pat1, pat2))

    # Apply Bonferroni correction to adjust p-values for multiple comparisons
    adjusted_p_values = multipletests(p_values, method='bonferroni')[1]

    # Display significant pairwise comparisons after correction
    significant_pairs = [(comparisons[i], adjusted_p_values[i]) for i in range(len(adjusted_p_values)) if adjusted_p_values[i] < 0.05]

    if significant_pairs:
        print("\nSignificant Differences Between Patients (Mann-Whitney U Test with Bonferroni correction):")
        for pair, adj_p in significant_pairs:
            print(f"Patients: {pair[0]} vs {pair[1]}, Adjusted p-value: {adj_p}")
    else:
        print("\nNo statistically significant differences were found between patients.")
else:
    print("No statistically significant differences found between patient groups.")
```
